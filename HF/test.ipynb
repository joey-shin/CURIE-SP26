# H2 Calculations

## Parse Method

- Parse job configuration and atoms in input file and store information
- Here, a list of Atom objects are created. Each Atom object contains a list of AO objects (Contracted Guassian Functions) and each AO object contains a list of Primitive Gaussian Functions

```python
H2_input = 'test_input/STO3G/H2.in'

config, atoms = parse(H2_input)
print(atoms)
print(config)
```

## Integral Build

```python
S = build_S(atoms)
T = build_T(atoms)
V = build_V(atoms)
ERI = build_ERI(atoms)
print('S: ')
print(S)
print('\nT: ')
print(T)
print('\nV: ')
print(V)
print('\nERI: ')
print(ERI)
```

## SCF loop

```python
C, P, F = scf(atoms, verbose=True)
```

```python
print('C: ')
print(C)
print('\nP: ')
print(P)
print('\nF: ')
print(F)
```
